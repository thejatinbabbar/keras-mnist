# Neural Networks built with Keras for MNIST Classification

**This notebook was run on Kaggle using the dataset from the Kaggle Competitions Library for the Challenge "[Digit Recognizer](https://www.kaggle.com/c/digit-recognizer/)"**

## Using the Keras API from Tensorflow

```python
import pandas as pd
from tensorflow import keras
```

```python
print(keras.__version__)
```

## Importing Data from the Kaggle Competetions Library

```python
df_train = pd.read_csv('../input/digit-recognizer/train.csv')
df_test = pd.read_csv('../input/digit-recognizer/test.csv')
```

## Preprocessing the data

```python
x_train, y_train = df_train.loc[:, 'pixel0':], df_train.loc[:, 'label']
x_test = df_test.loc[:, :]
```

```python
def normalize(feat):
    feat = feat / 255
    return feat
```

```python
x_train = normalize(x_train)
x_test = normalize(x_test)
```

```python
x_train = x_train.values.reshape(-1, 28, 28, 1)
x_test = x_test.values.reshape(-1, 28, 28, 1)
```

```python
y_train = keras.utils.to_categorical(y_train)
```

## Creating a Validation set out of the Training data

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.1, random_state=101)
```

## Creating the CNN model

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten
```

```python
model = Sequential()
model.add(Conv2D(16, (3,3), activation='relu', padding='same', input_shape=(28, 28, 1)))
model.add(MaxPool2D((2,2), strides=2))
model.add(Conv2D(32, (3,3), activation='relu', padding='same'))
model.add(MaxPool2D((2,2), strides=2))
model.add(Conv2D(64, (3,3), activation='relu', padding='same'))
model.add(MaxPool2D((2,2), strides=2))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.summary()
```

```python
model.compile(optimizer=keras.optimizers.Adam(), metrics='accuracy', loss='categorical_crossentropy')
```

## Training the CNN Model

```python
model.fit(x_train, y_train, batch_size=128, epochs=50)
```

## Evaluating the model with the Validation set

```python
model.evaluate(x_val, y_val)
```

```python
y_test = model.predict(x_test)
y_test = keras.backend.argmax(y_test).numpy()
```

## Creating the csv file for submission

```python
sub1 = pd.DataFrame()
sub1['Label'] = y_test
```

```python
sub1.index += 1
sub1.index.name = 'ImageId'
```

```python
sub1.reset_index(inplace=True)
```
